# line plot

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd


var = [1,2,3,4,5,6,7]
var1 = [2,3,5,4,6,1,6]

x=pd.DataFrame({"var":var,"var1":var1})

sns.lineplot(x="var",y="var1",data=x)
plt.show()
```

```python
plt.plot(var,var1)
plt.show()
```

```python
y1 = sns.load_dataset("penguins").head(20)
y1
```

```python
sns.lineplot(x="bill_length_mm",y="flipper_length_mm",data=y1,hue ="sex",size=10,style="sex",palette="Accent",markers=["o",">"],dashes=False,legend=False
            )
plt.grid()
plt.title("seaborn")
plt.show()
```

# bar plot

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
var = sns.load_dataset("penguins")
var
```

```python
sns.set(style="darkgrid")
order_1=["Dream","Torgersen","Biscoe"]
sns.barplot(x="island",y="flipper_length_mm",data=var,hue="sex",order=order_1,hue_order=["Female","Male"],n_boot=5,palette="icefire"
           ,saturation = 50,errcolor="y",capsize=0.3,dodge=False,alpha=0.5)#errwidth=10
plt.show()
```

# scatter plot

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
var = sns.load_dataset("penguins")
var
```

```python
m ={"Male":"*","Female":"o"}
sns.scatterplot(x="bill_length_mm",y="bill_depth_mm",data=var,hue="sex",style="sex",size="sex",
                sizes=(100,40),palette="cividis_r",alpha=0.8,markers=m)
plt.show()
```

# heatmap

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
var = np.linspace(1,10,20).reshape(4,5)
var
```

```python
sns.heatmap(var,vmin=0,vmax=10,cmap="PuOr",annot=True)
plt.show()
```

```python
data = sns.load_dataset("anagrams")
x = data.drop(columns=["attnr"],axis=1).head(10)
x
```

```python
sns.heatmap(x,annot=True,linewidth=4,linecolor="y")
plt.show()
```

# violin plot

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
var = sns.load_dataset("tips")
var
```

```python
sns.violinplot(x="day",y="total_bill",data=var,linewidth=2,palette="bwr")
plt.show()
```

```python
sns.violinplot(x="time",y="total_bill",data=var,linewidth=2,palette="bwr",order=["Dinner","Lunch"],inner="quart")
plt.show()
```

```python
sns.violinplot(x="day",y="total_bill",data=var,hue ="sex",split=True,scale="width")
plt.show()
```
